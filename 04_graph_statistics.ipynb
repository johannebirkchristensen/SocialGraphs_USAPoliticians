### Basic statistics

This notebooks shows some basic statistics of the graph G_politicans

```python
import matplotlib.pyplot as plt 
import pickle
import networkx as nx
```

```python

# load graph object from file
G = pickle.load(open('G_politicans.pickle', 'rb'))
```

```python
print(f"Number of nodes: {len(G.nodes())}")
print(f"Number of edges: {len(G.edges())}")
print("\n")
print(f"Max degree: {max(dict(G.degree()).values())}")
print(f"name of vertex with max degree: {max(dict(G.degree()), key=dict(G.degree()).get)}")
print(f"Max out degree: {max(dict(G.out_degree()).values())}")
print(f"name of vertex with max out degree: {max(dict(G.degree()), key=dict(G.degree()).get)}")
print(f"Max in degree: {max(dict(G.in_degree()).values())}")
print(f"name of vertex with max in degree: {max(dict(G.degree()), key=dict(G.degree()).get)}")
print(f"Min degree: {min(dict(G.degree()).values())}")

print("\n")
print(f"Average degree: {sum(dict(G.degree()).values())/G.number_of_nodes()}")
```

```python
out_degrees = [G.out_degree(n) for n in G.nodes()]
plt.hist(out_degrees, bins=100)
plt.title("Out degree distribution")
plt.xlabel("Out degree")
plt.ylabel("Number of nodes (log)")
plt.yscale('log')
plt.show()

in_degrees = [G.in_degree(n) for n in G.nodes()]
plt.hist(in_degrees, bins=100)
plt.title("In degree distribution")
plt.xlabel("In degree")
plt.ylabel("Number of nodes (log)")
plt.yscale('log')
plt.show()
```
